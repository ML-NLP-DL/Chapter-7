## chapter 7. 앙상블 학습과 랜덤 포레스트

* 앙상블 학습?

    여러개의 예측기로부터 예측을 수집하는 방식으로 학습.  
    이러한 예측기를 앙상블이라고 한다.  
    

* 대표적인 앙상블 기법  
    1. 배깅  
    2. 부스팅  
    3. 스태킹  

### 7.1 투표 기반 분류기  



    * 약한 학습기를 생성 (*약한 학습기? 랜덤보다 조금 더 높은 성능을 내는 예측기)  
    
    * 다양한 약한 학습기를 투표하면 강한 학습기 생성가능  
    
    
### 1. 직접 투표 분류기
  
    

    * 직접투표 분류기가 효과를 발휘하는 이유?  

    
*   __대수의 법칙__  
       단, 각 분류기가 독립적이고, 오차에 상관관계가 없어야함

### 2. 간접투표 분류기



각 투표기의 확률을 받아 이를 평균내어 최종 분류

### 7.2 배깅과 페이스팅  

* 배깅과 페이스팅은 같은 알고리즘을 사용하고, 훈련셋의 서브셋을 무작위 구성하는 방법  


* 배깅 vs 페이스팅
    1. 배깅  
        훈련세트에서 중복 샘플링 가능  
    2. 페이스팅  
        훈련세트에서 중복을 허용하지 않고 샘플링하는 방식  


  

분류 : 최빈값  
회귀 : 평균값

### 7.2.2 obb평가  
BaggingClassifier이용  
훈련 후 남은 샘플을 통해 평가

### 7.4 랜덤 포레스트  
* 랜덤포레스트?
    배깅 또는 페이스팅을 적용한 의사결정나무(Decision Tree)의 앙상블
    의사결정나무와 달리, 특성을 무작위 추출해서 최적의 특성을 찾아냄
    
### 7.4.1 익스트림 랜덤 트리
    임계값을 계산하지 않고, 특성을 무작위 분할한 뒤에 최상의 분할을 선택함

### 7.5 부스팅  

* 부스팅?  
    부스팅은 약한 학습기를 여러개 연결하여, 강한 학습기를 만드는 앙상블 기법
  
  

### 7.5.1 아다부스트  

  

이전모델이 과소학습한 샘플에 가중치를 주는 방식  
잘못 분류된 훈련 샘플의 가중치를 높여, 다음 학습모델에 입력

### 7.5.2 그래디언트 부스팅  
  
* 그래디언트 부스팅?
    이전 예측기의 잔여 오차를 새로운 예측기에 학습하는 기법  
    

### 7.6 스태킹  
  
* 스태킹?  
    예측을 취합하는 대신, 취합하는 모델을 훈련시키는 방법
    
      
    
    각 모델에서 예측된 값을 새로운 훈련데이터로 생성, 예측에 이용함.  
    최종 모델은 각 예측된 값을 특성으로 사용하는 데이터를 이용해 모델을 구성하고, 최종 예측
